<h1> FOR LOOP TO PUT ALL HTML AND NAME, RATING, .... <h1>

```python
name = []
mileage = []
dealer_name = []
rating = []
rating_count = []
price = []
```

```python
for result in results:
    #name
    try:
        name.append(result.find('h2').get_text())
    except:
        name.append("n/a")

    # mileage
    try:
        mileage.append(result.find('div',{'class':'mileage'}).get_text())
    except:
        mileage.append("n/a")
        
    # dealer name
    try:
        dealer_name.append(result.find('div',{'class':'dealer-name'}).get_text().strip())
    except:
        dealer_name.append("n/a")
    
    
    # rating count
    try:
        rating.append(result.find('span', {'class':'sds-rating__count'}).get_text())
    except:
        rating.append("n/a")
    
    
    # rating
    try:
        rating_count.append(result.find('span', {'class': 'sds-rating__link'}).get_text())
    except:
        rating_count.append("n/a")
        
    # price
    try:
        price.append(result.find('span', {'class': 'primary-price'}).get_text())
    except:
        price.append("n/a")
```

<h1> Pandas DATAFRAME <h1>

```python
car_dealer = pd.DataFrame({'Name' : name, "Mileage": mileage, "Dealer Name" : dealer_name, 
                           "Rating": rating, "Rating Count": rating_count, "Price": price})
```

```python
name = []
mileage = []
dealer_name = []
rating = []
rating_count = []
price = []

for i in range(1,500):
    website = 'https://www.cars.com/shopping/results/?page=' + str(i) + '&page_size=20&dealer_id=&keyword=&list_price_max=&list_price_min=&makes[]=mercedes_benz&maximum_distance=all&mileage_max=&sort=best_match_desc&stock_type=new_cpo&year_max=&year_min=&zip='
    
    response = requests.get(website)
    
    soup = BeautifulSoup(response.content, 'html.parser')
    
    results = soup.find_all('div', {'class':'vehicle-card'})
    
    for result in results:
        #name
        try:
            name.append(result.find('h2').get_text())
        except:
            name.append(np.nan)

        # mileage
        try:
            mileage.append(result.find('div',{'class':'mileage'}).get_text())
        except:
            mileage.append(np.nan)
            
        # dealer name
        try:
            dealer_name.append(result.find('div',{'class':'dealer-name'}).get_text().strip())
        except:
            dealer_name.append(np.nan)
        
        
        # rating count
        try:
            rating.append(result.find('span', {'class':'sds-rating__count'}).get_text())
        except:
            rating.append(np.nan)
        
        
        # rating
        try:
            rating_count.append(result.find('span', {'class': 'sds-rating__link'}).get_text())
        except:
            rating_count.append(np.nan)
            
        # price
        try:
            price.append(result.find('span', {'class': 'primary-price'}).get_text())
        except:
            price.append(np.nan)
    
```

```python
car_dealer = pd.DataFrame({'Name' : name, "Mileage": mileage, "Dealer Name" : dealer_name, 
                           "Rating": rating, "Rating Count": rating_count, "Price": price})
```
